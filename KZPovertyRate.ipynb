```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# New Antecedent/Consequent objects hold universe variables and membership
# functions
salary = ctrl.Antecedent(np.arange(29000, 163000, 1000), 'salary')
nchildren = ctrl.Antecedent(np.arange(0, 5, 1), 'nchildren')

salaryinstate = ctrl.Antecedent(np.arange(29000,163000, 1000), 'salaryinstate')

povertyRate = ctrl.Consequent(np.arange(0, 10, 1), 'povertyRate')

salary['poor'] = fuzz.trimf(salary.universe, [29000, 29000, 96000])
salary['mediocre'] = fuzz.trimf(salary.universe, [29000,96000,163000])
salary['medium'] = fuzz.trimf(salary.universe, [96000, 163000, 163000])



%matplotlib inline
salary.view(figsize = (20,20))
```
